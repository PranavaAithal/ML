```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("Food.csv",header=None)
```

```python
X=df[0]
Y=df[1]
```

```python
meanx = sum(X) / len(X)
meany = sum(Y) / len(Y)
num=0
den=0

for i in range(len(X)):
    num += (X[i] - meanx) * (Y[i] - meany)
    den += (X[i] - meanx) ** 2
```

```python
slope = num / den
intercept = meany - slope * meanx
```

```python
print("Linear Regression Equation : y = {:.2f}x + {:.2f}".format(slope, intercept))
```

```python
regline = [ slope * xi + intercept for xi in X ]
plt.scatter(X, Y, label = 'Data Points')
plt.plot(X, regline, color = 'green', label = 'Regression Line')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
df['y_pred']=slope*df[0]+intercept
df['error']=df[1]-df['y_pred']
df['error^2']=df['error']**2
df
```

```python
sse=df['error^2'].sum()
sse
```
